```python
import matplotlib.pyplot as plt
# Winkel für die Ecken des Quadrats (Draufsicht des Würfels)
winkel_quadrat = np.linspace(0, 2 * np.pi, 4, endpoint=False)

# Radien für die Ecken des Quadrats
radius_quadrat = np.ones_like(winkel_quadrat) * 5  # Verwenden denselben Radius wie für den Zylinder

# Winkel und Radien für die Verbindung der Ecken
winkel_verbindung = np.append(winkel_quadrat, winkel_quadrat[0])  # Schließt das Quadrat, indem der erste Punkt wiederholt wird
radius_verbindung = np.append(radius_quadrat, radius_quadrat[0])

# Polarplot für die Draufsicht des Würfels erstellen
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(winkel_verbindung, radius_verbindung)

# Anpassen des Plots
ax.set_ylim(0, 5 + 1)  # Etwas Raum um das Quadrat lassen
ax.set_yticklabels([])  # Entfernen der Radius-Beschriftungen
ax.set_xticklabels([])  # Entfernen der Winkel-Beschriftungen

plt.show()
```
